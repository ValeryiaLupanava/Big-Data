```python
#!/opt/anaconda/envs/bd9/bin/python

from urllib.parse import urlparse, unquote
import happybase
import os
import sys
import re
```

```python
import pandas as pd
import numpy as np
```

```python
def url2domain(url):
    try:
        a = urlparse(unquote(url.strip()))
        if (a.scheme in ['http','https']):
            b = re.search("(?:www\.)?(.*)",a.netloc).group(1)
            if b is not None:
                return str(b).strip()
            else:
                return ''
        else:
            return ''
    except:
        return ''
```

```python
# User categories
category_1 = 'Autouser'
category_2 = 'Businessuser'
category_3 = 'Homeuser'
category_4 = 'Bookuser'
# Analysed domains 
domains_category_1 = [u'cars.ru', u'avto-russia.ru', u'bmwclub.ru']
domains_category_2 = [u'zakon.kz', u'egov.kz', u'makler.md']
domains_category_3 = [u'russianfood.com', u'psychologies.ru', u'gotovim.ru']
domains_category_4 = [u'books.imhonet.ru', u'zhurnaly.biz', u'zvukobook.ru']

domains = [u'cars.ru', u'avto-russia.ru', u'bmwclub.ru', u'zakon.kz', u'egov.kz', u'makler.md', u'russianfood.com', u'psychologies.ru', u'gotovim.ru', u'books.imhonet.ru', u'zhurnaly.biz', u'zvukobook.ru']

# User identificators by categories
id1 = 2
id2 = 3
id3 = 4
id4 = 5
```

#### Maper saves only rows with domain of categories

```python
def mapper(line, data_holder):
    uid, ts, url = line.split('\t')
    if uid is not None and url is not None and (re.match(r'^http.+$', url) or re.match(r'^https.+$', url)):
        domain = url2domain(url)
        if domain in domains:
            data_holder.append([uid, ts, domain, url])
```

#### "for" cycle through all the data

```python
data_holder = []
for file in os.listdir('lab03/lab03data/'):
    with open(os.path.join('lab03/lab03data/', file), "r") as f:
        content = f.read()
    content_lines = content.splitlines()
    print(file)
    for line in content_lines:
        try:
            mapper(line, data_holder)
        except ValueError:
            continue
```

#### Convert data to pandas dataframe

```python
dataset = pd.DataFrame(data_holder)
```
